## Perceptron

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]

print(or_data)
```

```python
and_data['input1'] = [1,1,0,0]
and_data['input2'] = [1,0,1,0]
and_data['output'] = [1,0,0,0]

print(and_data)
```

```python
xor_data['input1'] = [1,1,0,0]
xor_data['input2'] = [1,0,1,0]
xor_data['output'] = [0,1,1,0]

print(and_data)
```

```python
## Create the visualization of all the above mentioned ge
```

```python
sns.scatterplot(x = xor_data['input1'], y=or_data['input2'], hue=or_data['output'], s = 200)
```

```python
sns.scatterplot(x = xor_data['input1'], y=xor_data['input2'], hue=xor_data['output'], s = 200)
```

```python
from sklearn.linear_model import Perceptron
or_classifiers = Perceptron()
and_classifiers = Perceptron()
xor_classifier = Perceptron()
```

```python
or_classifiers.fit(or_data.iloc[:, 0:2].values, or_data.iloc[:, -1].values)
and_classifiers.fit(and_data.iloc[:, 0:2].values, and_data.iloc[:, -1].values)
xor_classifier.fit(xor_data.iloc[:, 0:2].values, xor_data.iloc[:, -1].values)
```

```python
or_classifiers.coef_
```

```python
and_classifiers.coef_
```

```python
# Limitation of Perceptron. it is early able to create the linear decision boundaboundaries but non linear funtion. it want   be able to create a boundary 
```

```python
from mlxtend.plotting import plot_decision_regions

# set up subplots for side by side camparision
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# AND Classifiers Decision Boundary
axes[0].set_title('AND Gate Decision  Boundary')
plot_decision_regions(x = and_data.iloc[:, 0:2].values, y = and_data.iloc[:, -1].values, clf = and_classifiers, legend=2, ax=axes[0])
axes[0].set_xlabel['input1']
axes[0].set_xlabel['input2']

# OR Classifiers Decision Boundary
axes[0].set_title('OR Gate Decision  Boundary')
plot_decision_regions(x = or_data.iloc[:, 0:2].values, y = or_data.iloc[:, -1].values, clf = or_classifiers, legend=2, ax=axes[1])
axes[0].set_xlabel['input1']
axes[0].set_xlabel['input2']

# XOR Classifiers Decision Boundary
axes[0].set_title('XOR Gate Decision  Boundary')
plot_decision_regions(x = xor_data.iloc[:, 0:2].values, y = xor_data.iloc[:, -1].values, clf = xor_classifiers, legend=2, ax=axes[2])
axes[0].set_xlabel['input1']
axes[0].set_xlabel['input2']

plt.tight_layout()
plt.show()
```
