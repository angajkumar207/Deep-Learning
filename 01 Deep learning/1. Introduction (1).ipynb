# *Deep Learning*

Deep Learning is a subset of Machine Learning where computers learn using neural networks with many layers

### Simple vibe explanation

- Input layer >> takes raw data (image, text, audio)
- Hidden layers >> understand patterns (edges >> shapes >> objects)
- Output layer >> gives final answer (cat/dog, spam/not spam)
- The more layers = the deeper the learning = smarter pattern recognition

### Types of Deep Learning Models
1.  Feedforward Neural Network (FNN)
- - Data flows in one direction
  - Used in basic prediction tasks
2.  Convolutional Neural Network (CNN)
- - Used for images and videos
  - Detects patterns like edges, shapes
  - - Example: Face recognition
3.  Recurrent Neural Network (RNN)
- - Used for sequential data
  - Has memory of previous inputs
  - - Example: Language translation
4. Transformers
 - Modern architecture for NLP
  - Used in models like ChatGPT

### How Deep Learning Works
#### Step-by-step:
 1. Input data (image/text)
 2. Forward propagation
 3. Loss calculation (error)
 4. Backpropagation (error correction)
 5. Update weights

### Activation Functions
- ReLU (most used)
- Sigmoid
- Tanh

### Training Process
- **Forward Propagation**  → data passes through network
- **Loss Function** → measures error
- **Backpropagation** → adjusts weights
- **Optimization** → improves accuracy
### What Deep Learning is amazing at

- **Image recognition** >> Face unlock, medical scans
- **Speech recognition** >> Google Assistant, Siri
- **Text understanding** >> Chatbots, translation
- **Self-driving cars** >> Detecting roads, people, signs

### ML vs DL

| Machine Learning               | Deep Learning                 |
| ------------------------------ | ----------------------------- |
| Needs manual feature selection | Learns features automatically |
| Works with less data           | Needs **huge data**           |
| Simpler models                 | Very complex neural networks  |
