### Image Segmentation

Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Load image

```python
Image = cv2.imread("/content/birds for pro.jpg")
Image
```

```python
Image.shape
```

Convert the RGB image to Grayscale Image

```python
from skimage.color import rgb2gray
gray_image = rgb2gray(Image)
plt.imshow(gray_image, cmap='gray')
#


```

```python
gray_image.shape
```

```python
gray_r = gray_image.reshape(gray_image.shape[0] * gray_image.shape[1])
gray_r.shape
```

## Segmantic Segmentation: Two Pixel Values one for the object and another for the background

```python
for i in range(gray_r.shape[0]):
  if gray_r[i] > gray_r.mean():
    # object -> White
    gray_r[i] = 1
  else:
    # background ->>  black
    gray_r[i] = 0

gray_semantic_segmentation = gray_r.reshape(gray_image.shape[0], gray_image.shape[1])
plt.imshow(gray_semantic_segmentation, cmap='gray')
#
```

```python
gray_image = rgb2gray(Image)
gray_r = gray_image.reshape(gray_image.shape[0] * gray_image.shape[1])
gray_r
```

```python
for i in range(gray_r.shape[0]):
  # gray _r.mean() -> we have chose as the threshould
  if gray_r[i] > gray_r.mean():
    # object -> White
    gray_r[i] = 2

  # 0.5 as the Threshould
  elif gray_r[i] > 0.5:
    gray_r[i] =2
  # 0.25 as the threshould
  elif gray_r[i] > 0.25:
    gray_r[i] = 1
  else:
    # background ->>  blackS
    gray_r[i] = 0

gray_instance_segmentation = gray_r.reshape(gray_image.shape[0], gray_image.shape[1])
plt.imshow(gray_instance_segmentation, cmap='gray')
#
```
