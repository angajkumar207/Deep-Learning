## LeNet Practical in Keras
 ***LeNet*** : LeNet is one of the earliest and most influential Convolutional Neural Networks (CNNs), developed by Yann LeCun and colleagues in the 1990s for handwritten digit recognition.

#### What was it used for?

- LeNet was designed to recognize handwritten digits, such as those found on:

  - Bank checks
  - Postal ZIP codes
  - Forms

- LeNet-5 Architecture
  - The most famous version is LeNet-5.

##### Typical structure:
  - Input Image (32×32)
```text
 Input Image (32×32)
       ↓
 Convolution Layer
       ↓
 Average Pooling Layer
       ↓
 Convolution Layer
       ↓
 Average Pooling Layer
       ↓
 Fully Connected Layer
       ↓
 Fully Connected Layer
       ↓
```

### 1. import Libraries

````python
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.keras.layers import Conv2D
from keras.layers import Dense, Flatten, MaxPooling2D
from keras.models import Sequential
````

### 2. Loading the dataset and perfrom splitting

````python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
## Performimg reshaping operation
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
````

````python
x_train.shape[0]
````

````python
x_test.shape[0]
````

### 3. Normalization

````python
x_train = x_train / 255
x_test = x_test / 255
````

### 4. One Hot Encoding the Labels

````python
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)
````

### 5. Network Stucture

````python
model = Sequential()

model.add(Conv2D(6, kernel_size=(5,5), activation='relu', input_shape=(28, 28, 1))) ## 24 x  24 x 6
model.add(MaxPooling2D(pool_size=(2, 2))) # 12 X 12 X 6
model.add(Conv2D(16, kernel_size=(5, 5), activation='relu')) ##  8 x   8 x 16
model.add(MaxPooling2D(pool_size=(2, 2))) ## 4 x 4 x 16
model.add(Conv2D(120, kernel_size=(5, 5), activation='relu', padding='same')) ## 4 x 4 x 120
model.add(Flatten())
model.add(Dense(84, activation='relu'))
model.add(Dense(10, activation='softmax'))
````

````python
model.summary()
````

### 6. Model Compilation

````python
model.compile(loss=keras.metrics.categorical_crossentropy, optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))
````

### 7. Results

````python
score = model.evaluate(x_test, y_test)
print("Test loss: ", score[0])
print("Test Accuracy", score[1])
````
